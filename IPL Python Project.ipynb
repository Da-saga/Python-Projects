```python
# lets see how many team win the match by doing field and win the toss too
fields_win=ipl[(ipl['toss_decision'] == 'field') & (ipl['winner'] == ipl['toss_winner'])]
```

```python
#lets see winner team win the match by doing field and win the toss too
Fieldwin=fields_win['winner'].value_counts()
Fieldwin
```

```python
#now lets plot the bar graph
Fieldwin=fields_win['winner'].value_counts()
Fteam=list(Fieldwin.keys())
Fwin= list(Fieldwin)

plt.figure(figsize=(5,4))
plt.title('Win the Toss , Win the Match by doing Field',size=10)
plt.xticks(rotation=90,fontsize='x-small')
sns.barplot(x=Fteam,y=Fwin,edgecolor='black')
plt.xlabel('Team')
plt.ylabel('Win Counts')
plt.show()
```

```python
#now lets find out the win the toss,win the match by doing bat
BatWin=ipl[(ipl['toss_decision']=='bat') & (ipl['winner']==ipl['toss_winner'])]
Bat_win=BatWin['winner'].value_counts()
Bat_win
```

```python
#lets plot bar plot who win the toss,win the match by doing bat
Bat_win=BatWin['winner'].value_counts()
BatTeam=list(Bat_win.keys())
Batwin =list(Bat_win)

plt.figure(figsize=(5,4))
plt.title('Win the Toss,Win the Match by Batting')
plt.xticks(rotation=90,fontsize='x-small')
sns.barplot(x=BatTeam,y=Batwin,edgecolor='black')
plt.xlabel('Team')
plt.ylabel('Win Count')
plt.show()
```
